We find the number of bookings per hotel type, city hotels are most booked almost twice as compared to resort hotels.

```python
counts = data.hotel.value_counts()
print(counts)
```

We find the number of bookings per year , most bookings were done in 2016 , 2015 with the least number of bookings.

```python
year = data.arrival_date_year.unique()
```

```python
year_count = data.arrival_date_year.value_counts()
year_count
```

```python
plt.bar(year,year_count)
plt.xlabel('YEAR')
plt.ylabel('Number of bookings')
plt.show;
```

```python
months = data.arrival_date_month.unique()
months
```

We find number of bokings per month , it is expected that December has the highets but surprisingly it is August.

```python
bookings_per_month = data.arrival_date_month.value_counts()
bookings_per_month 
```

```python
plt.bar(months,bookings_per_month)
plt.xlabel('Months')
plt.ylabel('Bookings per month')
plt.title("Bookings Per Month")
plt.xticks(rotation=90)
plt.show;
```
